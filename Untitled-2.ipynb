```python
import pandas as pd

# Cargar los datos correctamente
csv_path = r"C:\Users\jotik\Visual Studio\Python\Educacion\Educacion.CSV"
df = pd.read_csv(csv_path, encoding='latin1', sep=';')

# Mostrar las primeras 5 filas
print(df.head())

# Información general
print(df.info())

# Ver cuántas filas y columnas tiene
print(f"Filas: {df.shape[0]}, Columnas: {df.shape[1]}")
```

```python
# Ver tipos de datos
print(df.dtypes)

# Identificar variables numéricas (cuantitativas)
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
print("Variables Cuantitativas:", num_cols)

# Identificar variables categóricas
cat_cols = df.select_dtypes(include=['object']).columns
print("Variables Categóricas:", cat_cols)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
df["DIRECTORIO"].value_counts().nlargest(30).plot(kind='bar')  # Muestra solo los 30 más comunes
plt.xticks(rotation=90)
plt.show()
```

```python
empty_cols = [col for col in num_cols if df[col].nunique() == 0]
print("Columnas vacías:", empty_cols)
```

```python
df = df.drop(columns=empty_cols)
```

```python
print("Cantidad de filas después de limpieza:", df.shape[0])
print("Valores NaN restantes por columna:")
print(df.isna().sum())
```

```python
for col in num_cols:
    unique_values = df[col].nunique()
    print(f"Procesando {col}... {unique_values} valores únicos")

    if unique_values > 1:
        plt.figure(figsize=(6, 4))
        sns.boxplot(x=df[col])
        plt.title(f"Boxplot de {col}")
        plt.show()
    else:
        print(f"La columna {col} se omitió porque tiene valores constantes o está vacía.")
```
